## Dependence of the SVD on geometry

```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
n = 1000
q = int(n / 4)
X = np.zeros((n, n))
X[(q - 1) : (3 * q), (q - 1) : (3 * q)] = 1
```

```python
n = 1000
q = int(n / 4)
X = np.zeros((n, n))
X[(q - 1) : (3 * q), (q - 1) : (3 * q)] = 1

X_rot = skimage.transform.rotate(X, 10)
X_rot[np.nonzero(X_rot)] = 1

fig, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(X, cmap="gray")
ax[0].set_title("Unrotated Matrix")

ax[1].imshow(X_rot, cmap="gray")
ax[1].set_title("Rotated Matrix")

plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=0)


fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].semilogy(S, "-o", color="k")
ax[0].set_ylim((10 ** (-16), 10 ** (4) + 1))
ax[0].set_xlim([0, 1000])
ax[0].grid()
ax[0].set_title("Unrotated Matrix: Spectrum")


U_rot, S_rot, VT_rot = np.linalg.svd(X_rot, full_matrices=0)

ax[1].semilogy(S_rot, "-o", color="k")
ax[1].set_ylim((10 ** (-16), 10 ** (4) + 1))
ax[1].set_xlim([0, 1000])
ax[1].grid()
ax[1].set_title("Rotated Matrix: Spectrum")

plt.show()
```
